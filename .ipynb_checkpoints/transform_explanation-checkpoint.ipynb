```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sales_transactions.csv')
```

```python
df
```

```python
#QUESTION: What percentage of total sales does each order represent? 
df.groupby('order')['ext price'].sum().transform(lambda x: x/(df['ext price'].sum()))
```

```python
#Question: What percentage of each each orders revenue does each sku represent? 

#approach w/o using transform
df['order_total']=df['order'].map(df.groupby('order')['ext price'].sum())
df.insert(5,'order_per',df['ext price']/df['order_total'])
```

```python
df
```

```python
#approach w/ transform
df=df.join(df.groupby('order')["ext price"].transform('sum').rename('order_total'))
df['order_per']=df['ext price']/df['order_total']
```

```python
#or...
df['order_per'] = df['ext price']/df.groupby('order')["ext price"].transform('sum')
```
